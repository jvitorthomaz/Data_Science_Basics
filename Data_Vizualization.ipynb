```python
import pandas as pd
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/alura-cursos/bibliotecas_visualizacao/main/Dados/imigrantes_canada.csv')
```

```python
df.set_index('País', inplace=True)
```

```python
anos = list(map(str, range(1980, 2014)))
```

```python
anos
```

```python
brasil = df.loc['Brasil', anos]
```

```python
brasil
```

```python
brasil_dict = {
    'ano': brasil.index.tolist(),
    'imigrantes': brasil.values.tolist()
}
dados_brasil = pd.DataFrame(brasil_dict)
```

```python
cores = []

for pais in america_sul_sorted.index:
  if pais == 'Brasil':
    cores.append('green')
  else:
    cores.append('silver')

fig, ax = plt.subplots(figsize=(12,5))
ax.barh(america_sul_sorted.index, america_sul_sorted['Total'], color=cores)
ax.set_title('América do Sul: Brasil foi o quarto país com mais imigrantes \npara o Canadá no período de 1980 a 2013', fontsize=18, loc='left')
ax.set_xlabel('Número de imigrantes', fontsize=14)
ax.set_ylabel('', fontsize=14)
ax.xaxis.set_tick_params(labelsize=12)
ax.yaxis.set_tick_params(labelsize=12)

# Dentro do text():
# O primeiro parametro v + 20, indicando a posição onde o texto será adicionado, onde o v corresponde a cada uma das barras.
# O i é a posição Y do nosso texto. Por fim, o str() para indicar que irá converter o tamanho de cada barra em uma string,
# ha e va representam o alinhamento horizontal e vertical
for i, v in enumerate(america_sul_sorted['Total']):
  ax.text(v + 20, i, str(v), color='black', fontsize=10, ha='left', va= 'center')

plt.show()
```

```python
cores = []

for pais in america_sul_sorted.index:
  if pais == 'Brasil':
    cores.append('green')
  else:
    cores.append('silver')

fig, ax = plt.subplots(figsize=(12,5))
ax.barh(america_sul_sorted.index, america_sul_sorted['Total'], color=cores)
ax.set_title('América do Sul: Brasil foi o quarto país com mais imigrantes \npara o Canadá no período de 1980 a 2013', fontsize=18, loc='left')
ax.set_xlabel('Número de imigrantes', fontsize=14)
ax.set_ylabel('', fontsize=14)
ax.xaxis.set_tick_params(labelsize=12)
ax.yaxis.set_tick_params(labelsize=12)

# Dentro do text():
# O primeiro parametro v + 20, indicando a posição onde o texto será adicionado, onde o v corresponde a cada uma das barras.
# O i é a posição Y do nosso texto. Por fim, o str() para indicar que irá converter o tamanho de cada barra em uma string,
# ha e va representam o alinhamento horizontal e vertical
for i, v in enumerate(america_sul_sorted['Total']):
  ax.text(v + 20, i, str(v), color='black', fontsize=10, ha='left', va= 'center')


ax.set_frame_on(False)
ax.get_xaxis().set_visible(False)
ax.tick_params(axis='both', which='both', length=0)

fig.savefig('imigracao_america_sul_canada.png', transparent=False, dpi=300, bbox_inches='tight')

plt.show()
```

```python
def gerar_grafico_paleta(palette):
    fig, ax = plt.subplots(figsize=(8,4))
    ax = sns.barplot(data=top_10, y=top_10.index, x='Total', orient='h', palette=palette)
    ax.set_title('Países com maior imigração para o Canadá\n1980 a 2013', loc='left', fontsize=18)
    ax.set_xlabel('Número de imigrantes', fontsize=14)
    ax.set_ylabel('')
    plt.show()
```
